```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile
file_path = 'E:\\Phd\\Project_Landfill_fires\\Data\\2024\\Ghazipur\\India_Shape\\india_st.shp'
gdf = gpd.read_file(file_path)

# Plot the entire map with gray color
fig, ax = plt.subplots(figsize=(12, 10))
gdf.plot(ax=ax, color='lightgray', edgecolor='black')

# Highlight Delhi (assuming 'Delhi' is a value in the geometry)
# Replace 'Delhi' with the appropriate name used in your shapefile
delhi = gdf[gdf['STATE'] == 'Delhi']  # Change 'NAME' to the actual column name for state/city names
delhi.plot(ax=ax, color='orange', edgecolor='black')

# Set the title and labels
ax.set_title('Map of India with Delhi Highlighted', fontsize=15)
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the plot
plt.show()
```
